```python
#3 determinar que criptos estan por encima de la media

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv('/content/drive/MyDrive/algorand.csv')
df2 = pd.read_csv('/content/drive/MyDrive/apecoin.csv')
df3 = pd.read_csv('/content/drive/MyDrive/binancecoin.csv')
df4 = pd.read_csv('/content/drive/MyDrive/bitcoin-cash.csv')

df1['date'] = pd.to_datetime(df1['date'])
df2['date'] = pd.to_datetime(df2['date'])
df3['date'] = pd.to_datetime(df3['date'])
df4['date'] = pd.to_datetime(df4['date'])

fecha1 = '2023-01-01'
fecha2 = '2023-12-31'

rango_fechas_df1 = df1.loc[(df1['date'] >= fecha1) & (df1['date'] <= fecha2)]
rango_fechas_df2 = df2.loc[(df2['date'] >= fecha1) & (df2['date'] <= fecha2)]
rango_fechas_df3 = df3.loc[(df3['date'] >= fecha1) & (df3['date'] <= fecha2)]
rango_fechas_df4 = df4.loc[(df4['date'] >= fecha1) & (df4['date'] <= fecha2)]

media_algorand = rango_fechas_df1['price'].mean()
media_apecoin = rango_fechas_df2['price'].mean()
media_binancecoin = rango_fechas_df3['price'].mean()
media_bitcoincash = rango_fechas_df4['price'].mean()

superiores_algorand = rango_fechas_df1[rango_fechas_df1['price'] > media_algorand]
superiores_apecoin = rango_fechas_df2[rango_fechas_df2['price'] > media_apecoin]
superiores_binancecoin = rango_fechas_df3[rango_fechas_df3['price'] > media_binancecoin]
superiores_bitcoincash = rango_fechas_df4[rango_fechas_df4['price'] > media_bitcoincash]

fig, axs = plt.subplots(2, 2, figsize=(15, 10))

axs[0, 0].plot(rango_fechas_df1['date'], rango_fechas_df1['price'], label='Precio')
axs[0, 0].plot(superiores_algorand['date'], superiores_algorand['price'], 'o', label='Por encima de la media', color='green')
axs[0, 0].axhline(media_algorand, color='red', linestyle='--', label='Media')
axs[0, 0].set_title('Algorand')
axs[0, 0].set_xlabel('Fecha')
axs[0, 0].set_ylabel('Precio')
axs[0, 0].legend()

axs[0, 1].plot(rango_fechas_df2['date'], rango_fechas_df2['price'], label='Precio')
axs[0, 1].plot(superiores_apecoin['date'], superiores_apecoin['price'], 'o', label='Por encima de la media', color='green')
axs[0, 1].axhline(media_apecoin, color='red', linestyle='--', label='Media')
axs[0, 1].set_title('ApeCoin')
axs[0, 1].set_xlabel('Fecha')
axs[0, 1].set_ylabel('Precio')
axs[0, 1].legend()

axs[1, 0].plot(rango_fechas_df3['date'], rango_fechas_df3['price'], label='Precio')
axs[1, 0].plot(superiores_binancecoin['date'], superiores_binancecoin['price'], 'o', label='Por encima de la media', color='green')
axs[1, 0].axhline(media_binancecoin, color='red', linestyle='--', label='Media')
axs[1, 0].set_title('BinanceCoin')
axs[1, 0].set_xlabel('Fecha')
axs[1, 0].set_ylabel('Precio')
axs[1, 0].legend()

axs[1, 1].plot(rango_fechas_df4['date'], rango_fechas_df4['price'], label='Precio')
axs[1, 1].plot(superiores_bitcoincash['date'], superiores_bitcoincash['price'], 'o', label='Por encima de la media', color='green')
axs[1, 1].axhline(media_bitcoincash, color='red', linestyle='--', label='Media')
axs[1, 1].set_title('Bitcoin Cash')
axs[1, 1].set_xlabel('Fecha')
axs[1, 1].set_ylabel('Precio')
axs[1, 1].legend()

plt.tight_layout()
plt.show()

print("Criptomonedas con precios por encima de la media en 2023:")
print(f"Algorand: {len(superiores_algorand)} días")
print(f"ApeCoin: {len(superiores_apecoin)} días")
print(f"BinanceCoin: {len(superiores_binancecoin)} días")
print(f"Bitcoin Cash: {len(superiores_bitcoincash)} días")
```

```python
#punto 5

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv('/content/drive/MyDrive/algorand.csv')
df2 = pd.read_csv('/content/drive/MyDrive/apecoin.csv')
df3 = pd.read_csv('/content/drive/MyDrive/binancecoin.csv')
df4 = pd.read_csv('/content/drive/MyDrive/bitcoin-cash.csv')

df1['date'] = pd.to_datetime(df1['date'])
df2['date'] = pd.to_datetime(df2['date'])
df3['date'] = pd.to_datetime(df3['date'])
df4['date'] = pd.to_datetime(df4['date'])

fecha1 = '2023-01-01'
fecha2 = '2023-12-31'

rango_fechas_df1 = df1.loc[(df1['date'] >= fecha1) & (df1['date'] <= fecha2)]
rango_fechas_df2 = df2.loc[(df2['date'] >= fecha1) & (df2['date'] <= fecha2)]
rango_fechas_df3 = df3.loc[(df3['date'] >= fecha1) & (df3['date'] <= fecha2)]
rango_fechas_df4 = df4.loc[(df4['date'] >= fecha1) & (df4['date'] <= fecha2)]

rango_fechas_df1['daily_pc'] = rango_fechas_df1['price'].pct_change() * 100
rango_fechas_df2['daily_pc'] = rango_fechas_df2['price'].pct_change() * 100
rango_fechas_df3['daily_pc'] = rango_fechas_df3['price'].pct_change() * 100
rango_fechas_df4['daily_pc'] = rango_fechas_df4['price'].pct_change() * 100

fig, axs = plt.subplots(2, 2, figsize=(15, 10))

sns.histplot(rango_fechas_df1['daily_pc'].dropna(), kde=True, ax=axs[0, 0], color='blue')
axs[0, 0].set_title('Algorand - Cambio porcentual diario ')
axs[0, 0].set_xlabel('Cambio porcentual')
axs[0, 0].set_ylabel('Frequency')

sns.histplot(rango_fechas_df2['daily_pc'].dropna(), kde=True, ax=axs[0, 1], color='green')
axs[0, 1].set_title('ApeCoin - Cambio porcentual diario ')
axs[0, 1].set_xlabel('Cambio porcentual')
axs[0, 1].set_ylabel('Frequency')

sns.histplot(rango_fechas_df3['daily_pc'].dropna(), kde=True, ax=axs[1, 0], color='red')
axs[1, 0].set_title('BinanceCoin - Cambio porcentual diario ')
axs[1, 0].set_xlabel('Cambio porcentual')
axs[1, 0].set_ylabel('Frequency')

sns.histplot(rango_fechas_df4['daily_pc'].dropna(), kde=True, ax=axs[1, 1], color='purple')
axs[1, 1].set_title('Bitcoin Cash - Cambio porcentual diario ')
axs[1, 1].set_xlabel('Cambio porcentual')
axs[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv('/content/drive/MyDrive/algorand.csv')
df2 = pd.read_csv('/content/drive/MyDrive/apecoin.csv')
df3 = pd.read_csv('/content/drive/MyDrive/binancecoin.csv')
df4 = pd.read_csv('/content/drive/MyDrive/bitcoin-cash.csv')

df1['date'] = pd.to_datetime(df1['date'])
df2['date'] = pd.to_datetime(df2['date'])
df3['date'] = pd.to_datetime(df3['date'])
df4['date'] = pd.to_datetime(df4['date'])

fecha1 = '2023-01-01'
fecha2 = '2023-12-31'

rango_fechas_df1 = df1.loc[(df1['date'] >= fecha1) & (df1['date'] <= fecha2)]
rango_fechas_df2 = df2.loc[(df2['date'] >= fecha1) & (df2['date'] <= fecha2)]
rango_fechas_df3 = df3.loc[(df3['date'] >= fecha1) & (df3['date'] <= fecha2)]
rango_fechas_df4 = df4.loc[(df4['date'] >= fecha1) & (df4['date'] <= fecha2)]

rango_fechas_df1['daily_pc'] = rango_fechas_df1['price'].pct_change() * 100
rango_fechas_df2['daily_pc'] = rango_fechas_df2['price'].pct_change() * 100
rango_fechas_df3['daily_pc'] = rango_fechas_df3['price'].pct_change() * 100
rango_fechas_df4['daily_pc'] = rango_fechas_df4['price'].pct_change() * 100

fig, ax = plt.subplots(figsize=(12, 8))

data = [rango_fechas_df1['daily_pc'].dropna(), rango_fechas_df2['daily_pc'].dropna(), rango_fechas_df3['daily_pc'].dropna(), rango_fechas_df4['daily_pc'].dropna()]
labels = ['Algorand', 'ApeCoin', 'BinanceCoin', 'Bitcoin Cash']

ax.boxplot(data, labels=labels, patch_artist=True, boxprops=dict(facecolor='skyblue'))
ax.set_title('Distribución de Cambios Porcentuales Diarios en 2023')
ax.set_xlabel('Criptomonedas')
ax.set_ylabel('Cambios Porcentuales Diarios')

plt.show()
```
